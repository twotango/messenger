```python
import pandas as pd
```

```python
watch = pd.read_csv('/Users/emilkoch/Downloads/export-6.csv')
```

```python
watch
```

```python
# Parse values in 'values_1' column to extract numerical values
watch['values_1'] = watch['values_1'].apply(lambda x: [float(val.strip("[]'")) for val in x.split(", ")]) 
```

```python
# Convert the 'values_1' column to numeric if possible
watch['values_1'] = watch['values_1'].apply(pd.to_numeric, errors='coerce')

# Pivot the data to create columns for each unique source_encoded
pivot_data = pd.pivot_table(watch, 
                            values='values_1', 
                            index='timestamp', 
                            columns='source')

# Print the pivoted data
print(pivot_data)
```

```python
pivot_data 
```

```python
pivot_data4 = pivot_data.drop(columns = ['battery', 'heart_rate', 'pressure'])
```

```python
# Expand accelerometer column into x, y, z components
accelerometer_components = pivot_data4 ['accelerometer'].apply(pd.Series)
accelerometer_components.columns = ['accel_x', 'accel_y', 'accel_z']

# Expand gravity column into x, y, z components
gravity_components = pivot_data4 ['gravity'].apply(pd.Series)
gravity_components.columns = ['gravity_x', 'gravity_y', 'gravity_z']

# Expand gyroscope column into x, y, z components
gyroscope_components = pivot_data4 ['gyroscope'].apply(pd.Series)
gyroscope_components.columns = ['gyro_x', 'gyro_y', 'gyro_z']

# Expand linear_acceleration column into x, y, z components
linear_acceleration_components = pivot_data4 ['linear_acceleration'].apply(pd.Series)
linear_acceleration_components.columns = ['linear_accel_x', 'linear_accel_y', 'linear_accel_z']

# Expand magnetometer column into x, y, z components
magnetometer_components = pivot_data4 ['magnetometer'].apply(pd.Series)
magnetometer_components.columns = ['magnetometer_x', 'magnetometer_y', 'magnetometer_z']

# Expand orientation column into x, y, z components
orientation_components = pivot_data4['orientation'].apply(pd.Series)
orientation_components.columns = ['orientation_x', 'orientation_y', 'orientation_z']

rotation_components = pivot_data4['rotation_vector'].apply(pd.Series)
rotation_components.columns = ['rotation_scalar', 'rotation_i', 'rotation_j', 'rotation_k', 'rotation_angle']

# Concatenate the new components with the original DataFrame
pivot_data4  = pd.concat([pivot_data4 , accelerometer_components, gravity_components, gyroscope_components, linear_acceleration_components, magnetometer_components, orientation_components, rotation_components], axis=1)

# Drop the original columns
pivot_data4.drop(columns=['accelerometer', 'gravity', 'gyroscope', 'linear_acceleration', 'magnetometer', 'orientation', 'rotation_vector'], inplace=True)

# Print the updated DataFrame
print(pivot_data4)
```

```python
pivot_data4
```

```python
# Remove duplicate columns, if any
pivot_data4 = pivot_data4.loc[:, ~pivot_data4.columns.duplicated()]
# Replace NaN values with 0s
pivot_data4.fillna(0, inplace=True)
```

```python
# Convert the values in 'step_detector' column to strings
pivot_data4['step_detector'] = pivot_data4['step_detector'].astype(str)

# Remove square brackets and single quotes from the values in 'step_detector' column
pivot_data4['step_detector'] = pivot_data4['step_detector'].str.strip("[]").str.replace("'", "")

# Convert the values in 'step_detector' column to float
pivot_data4['step_detector'] = pivot_data4['step_detector'].astype(float)
```

```python
# Convert the values in 'step_detector' column to strings
pivot_data4['step_counter'] = pivot_data4['step_counter'].astype(str)

# Remove square brackets and single quotes from the values in 'step_detector' column
pivot_data4['step_counter'] = pivot_data4['step_counter'].str.strip("[]").str.replace("'", "")

# Convert the values in 'step_detector' column to float
pivot_data4['step_counter'] = pivot_data4['step_counter'].astype(float)
```
