I aim to combine activity_type and smartwatch_source data sets to potentially analyze upon activity_type. I will merge them grouped by time stamp. 

```python
# Convert timestamp columns to datetime objects
activity_type_full['from_1'] = pd.to_datetime(activity_type_full['from_1'])
smartwatch_sources['timestamp'] = pd.to_datetime(smartwatch_sources['timestamp'])

# Merge the data frames based on timestamps
merged_data = pd.merge_asof(smartwatch_sources.sort_values('timestamp'), 
                             activity_type_full.sort_values('from_1'),
                             left_on='timestamp',
                             right_on='from_1',
                             direction='backward')

# Optionally, drop redundant columns or rename them as needed
merged_data.drop(columns=['from_1', 'to_1'], inplace=True)
merged_data.rename(columns={'activity_type': 'activity'}, inplace=True)

# Fill NaN values with 0s
merged_data.fillna(0, inplace=True)

# Print or further process the merged data
print(merged_data)

```

I am replacing NaN with 0s and filtering rows where activity and duration are not equal 0. 

```python
# Filter the DataFrame to keep rows where both "activity" and "duration" are not equal to 0
filtered_data = merged_data[(merged_data['activity'] != 0) & (merged_data['duration'] != 0)]

# Print or further process the filtered data
print(filtered_data)
```

```python
filtered_data 
```

```python
import networkx as nx
import matplotlib.pyplot as plt

# Extract relevant columns: activity, timestamp, source
sensor_activity_data = filtered_data[['activity', 'timestamp', 'source']]

# Create a bipartite graph
B = nx.Graph()

# Add nodes for activities and sensor data
activities = sensor_activity_data['activity'].unique()
sensors = sensor_activity_data['source'].unique()
B.add_nodes_from(activities, bipartite=0)
B.add_nodes_from(sensors, bipartite=1)

# Add edges between activities and sensor data based on their occurrence together
for activity in activities:
    activity_data = sensor_activity_data[sensor_activity_data['activity'] == activity]
    for _, row in activity_data.iterrows():
        B.add_edge(activity, row['source'])

# Visualize the bipartite graph
pos = nx.bipartite_layout(B, activities)
nx.draw(B, pos, with_labels=True, node_size=3000, node_color='skyblue', font_size=10)
plt.title('Bipartite Graph of Activities and Sensor Data')
plt.show()
```

Given that the network analysis hasn't been truly informative, I will first look at clusters. For that purpose I will label encode the categories. 

```python
from sklearn.preprocessing import LabelEncoder

# Create a label encoder object
label_encoder = LabelEncoder()

# Apply label encoding to the "activity" column
filtered_data['activity_encoded'] = label_encoder.fit_transform(filtered_data['activity'])

# Apply label encoding to the "source" column
filtered_data['source_encoded'] = label_encoder.fit_transform(filtered_data['source'])

# Create dictionaries to map encoded values back to original categorical values
activity_encoder_dict = dict(zip(filtered_data['activity'], filtered_data['activity_encoded']))
source_encoder_dict = dict(zip(filtered_data['source'], filtered_data['source_encoded']))

# Display the dictionaries
print("Activity Encoding Dictionary:")
print(activity_encoder_dict)

print("\nSource Encoding Dictionary:")
print(source_encoder_dict)
```

Finally, use string manipulation methods like .strip(), .split(), and then convert each element to a float.

```python
# Convert 'timestamp' column to datetime format
filtered_filtered_data['timestamp'] = pd.to_datetime(filtered_filtered_data['timestamp'])

# Parse values in 'values_1' column to extract numerical values
filtered_filtered_data['values_1'] = filtered_filtered_data['values_1'].apply(lambda x: [float(val.strip("[]'")) for val in x.split(", ")])
```

```python
# Convert the 'values_1' column to numeric if possible
filtered_filtered_data['values_1'] = filtered_filtered_data['values_1'].apply(pd.to_numeric, errors='coerce')

# Pivot the data to create columns for each unique source_encoded
pivot_data = pd.pivot_table(filtered_filtered_data, 
                            values='values_1', 
                            index='timestamp', 
                            columns='source_encoded')

# Print the pivoted data
print(pivot_data)

```

```python
# Replace 'source_encoded' with 'source'
pivot_data.columns.name = 'source'

pivot_data
```

I have pivoted the data frame for better interpretability. Later, it would be more scaleable, though, having the more "concentrated" structure when dealing with large data sets. For research purposes, though, it's better to pivot the data columns. 

```python
# Group 'filtered_filtered_data' by timestamp and aggregate 'activity_encoded'
activity_grouped = filtered_filtered_data.groupby('timestamp')['activity_encoded'].first()

# Merge 'activity_grouped' with 'pivot_data' based on timestamp index
pivot_data = pivot_data.merge(activity_grouped, how='left', on='timestamp')

# Reset index to make timestamp a column again
pivot_data.reset_index(inplace=True)
```

```python
pivot_data
```

```python
# Group 'filtered_filtered_data' by both 'timestamp' and 'activity_encoded' and aggregate 'duration'
duration_grouped = filtered_data.groupby(['timestamp', 'activity_encoded'])['duration'].first()

# Merge 'duration_grouped' with 'pivot_data' based on timestamp index
pivot_data = pivot_data.merge(duration_grouped, how='left', on='timestamp')

# Print the updated pivoted data with 'duration' as a column specific to each activity
print(pivot_data)
```

Fill NaN with zeroes and get back the original column names. 

```python
# Replace the encoded column names with their original names
source_names = {0: 'accelerometer', 1: 'gravity', 2: 'gyroscope', 3: 'linear_acceleration'}  # Replace with actual source names
pivot_data.rename(columns=source_names, inplace=True)

# Replace NaN values with 0s
pivot_data.fillna(0, inplace=True)

# Print the updated pivoted data
print(pivot_data)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Convert timestamp column to datetime format
pivot_data['timestamp'] = pd.to_datetime(pivot_data['timestamp'])

# Convert any integer values in the 'accelerometer' column to lists
pivot_data['accelerometer'] = pivot_data['accelerometer'].apply(lambda x: [x] if isinstance(x, int) else x)

# Now, you can extract accelerometer readings for each axis
pivot_data['accel_x'] = pivot_data['accelerometer'].apply(lambda x: x[0] if len(x) > 0 else np.nan)
pivot_data['accel_y'] = pivot_data['accelerometer'].apply(lambda x: x[1] if len(x) > 1 else np.nan)
pivot_data['accel_z'] = pivot_data['accelerometer'].apply(lambda x: x[2] if len(x) > 2 else np.nan)

# Plot accelerometer data for each axis over time
plt.figure(figsize=(12, 6))
plt.plot(pivot_data['timestamp'], pivot_data['accel_x'], label='X-axis')
plt.plot(pivot_data['timestamp'], pivot_data['accel_y'], label='Y-axis')
plt.plot(pivot_data['timestamp'], pivot_data['accel_z'], label='Z-axis')
plt.title('Accelerometer Readings Over Time')
plt.xlabel('Timestamp')
plt.ylabel('Acceleration')
plt.legend()
plt.show()
```

```python
# Expand accelerometer column into x, y, z components
accelerometer_components = pivot_data['accelerometer'].apply(pd.Series)
accelerometer_components.columns = ['accel_x', 'accel_y', 'accel_z']

# Expand gravity column into x, y, z components
gravity_components = pivot_data['gravity'].apply(pd.Series)
gravity_components.columns = ['gravity_x', 'gravity_y', 'gravity_z']

# Expand gyroscope column into x, y, z components
gyroscope_components = pivot_data['gyroscope'].apply(pd.Series)
gyroscope_components.columns = ['gyro_x', 'gyro_y', 'gyro_z']

# Expand linear_acceleration column into x, y, z components
linear_acceleration_components = pivot_data['linear_acceleration'].apply(pd.Series)
linear_acceleration_components.columns = ['linear_accel_x', 'linear_accel_y', 'linear_accel_z']

# Concatenate the new components with the original DataFrame
pivot_data = pd.concat([pivot_data, accelerometer_components, gravity_components, gyroscope_components, linear_acceleration_components], axis=1)

# Drop the original columns
pivot_data.drop(columns=['accelerometer', 'gravity', 'gyroscope', 'linear_acceleration'], inplace=True)

# Print the updated DataFrame
print(pivot_data)
```
